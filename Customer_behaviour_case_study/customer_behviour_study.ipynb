```python
import matplotlib.pyplot as plt

gender_count = df["gender"].value_counts()
```

```python
plt.plot(gender_count,marker="o",linestyle="none")
plt.title("Gender wise count")
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
```

```python
# Customer average age
df["age"].mean().astype("int")
```

```python
# unique membership
df["membership_type"].unique()

```

```python
ms_count = df["membership_type"].value_counts()
```

```python
plt.plot(ms_count,marker="o",linestyle="dashed")
plt.xlabel("membership type",color="red")
plt.ylabel("customer count",color="red")
plt.title("membership wise count",color="green")
```

```python
# membership wise total_spend summary
membership_total_spend = df.groupby("membership_type")["total_spend"].sum()
membership_total_spend
```

```python
plt.bar(membership_total_spend.index,membership_total_spend.values,color="purple")
plt.xlabel("membership type")
plt.ylabel("membership count")
plt.title("membership wise chart",color = "red")
```

```python
# customers per city [plot]

customer_per_city = df["city"].value_counts()
customer_per_city
```

```python
plt.bar(customer_per_city.index,customer_per_city.values,color = "green")
```

```python
# city wise total spend [plot]
city_wise_total_spend = df.groupby("city")["total_spend"].sum()
city_wise_total_spend
```

```python
plt.bar(city_wise_total_spend.index,city_wise_total_spend.values,color="red",width=0.5)
plt.xlabel("CITY",color = "green")
plt.ylabel("COUNT",color = "green")
plt.title("MEMBERSHIP WISE COUNT",color = "green")
```

```python
plt.pie(city_wise_total_spend.values,labels=city_wise_total_spend.index)
```

```python
# city wise averae rating

city_wise_avg_rating = df.groupby("city")["average_rating"].mean()
city_wise_avg_rating
```

```python
plt.bar(city_wise_avg_rating.index,city_wise_avg_rating.values,width=0.5)
plt.xlabel("CITY",color="green")
plt.ylabel("AVERAGE RATING",color ="green")
plt.title("CITY WISE AVERAGE RATING")
```
