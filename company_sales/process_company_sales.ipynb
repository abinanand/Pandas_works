```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\ABIN ANAND\\Desktop\\ML_works\\pandas_works\\company_sales\\company_sales_data.csv")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#Total profit of all month

total_profit_month = df.groupby("month_number")["total_profit"].sum()
plt.plot(total_profit_month,marker = "o",color ="red",linestyle = "dotted")
plt.xlabel("MONTH")
plt.ylabel("TOTAL PROFIT")
plt.title("MONTH WISE TOTAL PROFIT")
```

```python
total_profit_unit = df.groupby("month_number")["total_units"].sum()
total_profit_unit
```

```python
plt.plot(total_profit_unit,marker="o",color="red",linestyle="dotted",linewidth =3)
plt.xlabel("MONTH NUMBER")
plt.ylabel("SOLID UNITS NUMBER")
plt.title("MONTH WISE TOTAL UNITS")
```

```python
products = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
for p in products:
    plt.plot(df["month_number"],df[p],marker="o",label=p)
plt.legend()
```

```python
plt.scatter(df["month_number"],df["toothpaste"],label="Toothpaste Sale")
plt.legend()
```

### Read face cream and facewash product sales data and show it using the bar chart with grid line

```python
bar_width = 0.35
plt.bar(df["month_number"]-bar_width/2,df["facecream"],width=bar_width,label="Faceream")
plt.bar(df["month_number"]+bar_width/2,df["facewash"],width=bar_width,label="Facewash")
plt.legend()
plt.grid()
```

### Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.hist(df["total_profit"],bins=10,edgecolor="white")
```

### Calculate total sale data for last year for each product and show it using a Pie chart

```python
products = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
Sale_data_last_year = {p:df[p].sum() for p in products}
plt.pie(Sale_data_last_year.values(),labels=products,autopct="%1.1f%%")
```

### Read Bathing soap facewash of all months and display it using the Subplot

```python
figure,axes = plt.subplots(1,2)

axes[0].plot(df["month_number"], df["bathingsoap"])
axes[0].set_title("Bathing Soap Sales")

axes[1].plot(df["month_number"], df["facewash"])
axes[1].set_title("Facewash Sales")

```
