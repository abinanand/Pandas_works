```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\ABIN ANAND\\Desktop\\ML_works\\pandas_works\\penguin_datasets\\data.csv")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df["bill_length_mm"].fillna(df["bill_length_mm"].mean(),inplace=True)
```

```python
df["bill_depth_mm"].fillna(df['bill_depth_mm'].mean(),inplace=True)
```

```python
df["flipper_length_mm"].fillna(df["flipper_length_mm"].mean(),inplace=True)
```

```python
df["body_mass_g"].fillna(df["body_mass_g"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["sex"].fillna(df["sex"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#flipper_length and body_mass

flip_length_body_mass = df.groupby("flipper_length_mm")["body_mass_g"].mean()
plt.plot(flip_length_body_mass,marker="o")
plt.xlabel("FLIPPER LENGTH")
plt.ylabel("BODY MASS")
plt.title("FLIPPER LENGTH VS BODY MASS")
```

```python
# bill_length and body mass
bill_length_body_mass = df.groupby("bill_length_mm")["body_mass_g"].mean()
plt.plot(bill_length_body_mass,marker="o")
plt.xlabel("BILL LENGTH")
plt.ylabel("BODY MASS")
plt.title("BILL LENGTH VS BODY MASS")
```

```python
# bill_depth and body mass
bill_depth_body_mass = df.groupby("bill_depth_mm")["body_mass_g"].mean()
plt.plot(bill_depth_body_mass,marker="o")
plt.xlabel("BILL DEPTH")
plt.ylabel("BODY MASS")
plt.title("BILL DEPTH VS BODY MASS")
```

```python
#Island wise body mass
island_wise_body_mass = df.groupby("island")["body_mass_g"].mean()
plt.bar(island_wise_body_mass.index,island_wise_body_mass.values)
plt.xlabel("ISLAND")
plt.ylabel("BODY MASS")
plt.title("ISLAND VS BODY MASS")
```

```python
#Island wise gender
island_wise_gender = df.groupby("island")["sex"].value_counts()
island_wise_gender
```

```python
# species wise body mass

species_body_mass = df.groupby("species")["body_mass_g"].mean()
species_body_mass
```

```python
plt.bar(species_body_mass.index,species_body_mass.values)
plt.xlabel("ISLAND")
plt.ylabel("BODY MASS")
plt.title("SPIECS VS BODY MASS")
```
